```python
#3.1  책 목록을 json 파일로 저장하기
import requests
import json
import os

NAVER_CLIENT_ID = "VNKujcDRjk_55Rq7DhD8"
NAVER_CLIENT_SECRET = "WuM8zeC4dp"

headers = {
    "X-Naver-Client-Id": NAVER_CLIENT_ID,
    "X-Naver-Client-Secret": NAVER_CLIENT_SECRET
}

def search_books(query):
    payload = {
        'query': query,
        'display': 50,
        'sort': 'sim'
    }
    url = "https://openapi.naver.com/v1/search/book.json"
    res = requests.get(url, headers=headers, params=payload)
    return res.json()["items"] if res.ok else []

def save_to_json(data, filename="data/books.json"):
    os.makedirs(os.path.dirname(filename), exist_ok=True)
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

query = "인공지능"  # 검색어
books = search_books(query)
save_to_json(books, "data/books.json")
print(" data/books.json 파일 저장 완료")
```

```python
#3.2 books.json 파일을 Pandas DataFrame로 저장하기
import pandas as pd
import json

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1000)

def load_json_to_df(filepath="data/books.json"):
    with open(filepath, "r", encoding="utf-8") as f:
        data = json.load(f)
    df = pd.DataFrame(data).reset_index(drop=True)
    df.index = df.index + 1
    return df

books_df = load_json_to_df("data/books.json")
print(f" DataFrame 로드 완료 (총 {len(books_df)}권)")
```

```python
#3.4 검색어로  찾은  책 목록 중에서 가격이 2만원 이상인 책만 출력하기
import pandas as pd

# 출력 옵션을 바꿔서 모든 행이 나오도록 설정
pd.set_option("display.max_rows", None)  

books_df["discount"] = pd.to_numeric(books_df["discount"], errors="coerce")

filtered_df = (
    books_df.loc[books_df["discount"] >= 20000, ["title", "author", "discount", "publisher", "pubdate"]]
    .sort_values(by="discount", ascending=False)
    .reset_index(drop=True)
)

filtered_df.index = filtered_df.index + 1  # 인덱스를 1부터 시작

display(filtered_df)
```
